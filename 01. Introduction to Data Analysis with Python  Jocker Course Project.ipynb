```python
# H1. 'Define a matrix whose rows correspond to countries and the columns to the regions.
#Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the country in row i is in the region in column j and a 0 otherwise.'

# Data pre processing for easy visual control on final result
data_Original.sort_values('Region', ascending=False, inplace=True)

# variables
row=data_Original['Country']; col=data_Original['Region'].unique();

# Create dictionnary for Country & Region
my_dictionnary=data_Original.set_index('Country')['Region'].to_dict()

# Initialization with array of zeros
Init=np.zeros(shape=(len(row),len(col))) # Init=np.zeros(shape=(len(row),len(data_Original['Region'].unique())))  Why this command does not work ?

# Create a DataFrame as a initialized matrix
dfMatrix= pd.DataFrame(Init, index=[row], columns=[col])

# Loop to browse the DataFrame and apply the test condition to fill in a 1 or not
#j=0
for i in range(0,dfMatrix.shape[1]):
    if j==dfMatrix.shape[0]: j=0;
    #print('**boucle i:**',' i:',i,' j:',j); # Loop process control
    for j in range(0,dfMatrix.shape[0]):
        #if j==0: print('**boucle j:**',' i:',i,' j:',j) # Loop process control
        if my_dictionnary[row[j]]==col[i]:
            dfMatrix.loc[row[j],col[i]]=1
dfMatrix
```
